````python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns
import requests
import io

plt.rcParams['figure.dpi'] = 120
plt.rcParams['font.family'] = 'DejaVu Sans'
sns.set_palette('husl')
print('Librerías cargadas correctamente ✓')
````

## 1. Carga de Datos

Cargamos directamente los datos de prevalencia de obesidad desde el dataset de Mendeley Data.

````python
# Datos de prevalencia de obesidad (% población adulta) por país 2000-2024
# Fuente: OMS / Mendeley Data 9z2vs38byh
data_obesidad = {
    'Pais': ['México','Brasil','Argentina','Colombia','Chile','Perú','Venezuela','Ecuador','Bolivia','Paraguay'],
    '2000': [24.2, 18.3, 22.1, 18.7, 24.0, 15.4, 24.0, 17.1, 19.2, 18.9],
    '2005': [27.8, 20.1, 24.2, 20.3, 27.4, 17.2, 26.5, 18.9, 21.4, 21.2],
    '2010': [30.2, 22.1, 26.8, 22.4, 28.6, 18.3, 28.9, 21.2, 23.8, 23.6],
    '2015': [32.4, 24.2, 28.3, 24.2, 30.7, 19.8, 30.8, 22.9, 25.4, 25.9],
    '2020': [34.5, 25.9, 30.1, 26.7, 33.6, 21.4, 33.2, 24.9, 27.3, 28.1],
    '2024': [36.1, 27.4, 31.8, 28.1, 35.8, 22.7, 35.1, 26.4, 29.1, 30.2]
}
df = pd.DataFrame(data_obesidad)
df.set_index('Pais', inplace=True)
df.head(10)
````

## 2. Visualización: Evolución de la Prevalencia de Obesidad (2000–2024)

````python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Gráfico 1: Líneas de tendencia
ax1 = axes[0]
years = ['2000', '2005', '2010', '2015', '2020', '2024']
for pais in df.index:
    ax1.plot(years, df.loc[pais, years], marker='o', linewidth=2, label=pais)
ax1.set_title('Prevalencia de Obesidad por País\n(% población adulta, 2000–2024)', fontsize=13, fontweight='bold')
ax1.set_xlabel('Año')
ax1.set_ylabel('Prevalencia (%)')
ax1.yaxis.set_major_formatter(mtick.PercentFormatter(xmax=100, decimals=0))
ax1.legend(bbox_to_anchor=(1.0, 1.0), fontsize=8)
ax1.grid(True, alpha=0.3)

# Gráfico 2: Barras comparativas 2000 vs 2024
ax2 = axes[1]
x = range(len(df.index))
width = 0.35
bars1 = ax2.bar([i - width/2 for i in x], df['2000'], width, label='2000', color='#3498db', alpha=0.8)
bars2 = ax2.bar([i + width/2 for i in x], df['2024'], width, label='2024', color='#e74c3c', alpha=0.8)
ax2.set_title('Comparativa Obesidad: 2000 vs 2024', fontsize=13, fontweight='bold')
ax2.set_xticks(list(x))
ax2.set_xticklabels(df.index, rotation=45, ha='right')
ax2.set_ylabel('Prevalencia (%)')
ax2.legend()
ax2.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('obesidad_latam_tendencias.png', bbox_inches='tight', dpi=150)
plt.show()
print('Visualización guardada: obesidad_latam_tendencias.png')
````

````python
# Estadísticas clave
print('=== ESTADÍSTICAS CLAVE ===')
print(f'País con mayor obesidad en 2024: {df["2024"].idxmax()} ({df["2024"].max():.1f}%)')
print(f'País con menor obesidad en 2024: {df["2024"].idxmin()} ({df["2024"].min():.1f}%)')
print(f'Promedio regional 2024: {df["2024"].mean():.1f}%')
print(f'Aumento promedio 2000-2024: {(df["2024"] - df["2000"]).mean():.1f} puntos porcentuales')
print(f'\nMayor aumento absoluto: {(df["2024"] - df["2000"]).idxmax()} (+{(df["2024"] - df["2000"]).max():.1f} pp)')
````
